```python
import pandas as pd

table = pd.read_csv('vgsale_1.csv', sep=',').drop_duplicates(['Name', 'Year'])
colors = ['lightpink', 'mediumaquamarine', 'peachpuff', 'powderblue']

# 1
count_2000 = table[table['Year'] <= 2000].groupby('Genre').agg({'Name' : 'count', 'Global_Sales' : 'sum'})
count_2000.columns = ['Number', 'Global_Sales']
count_2000['Sum'] = count_2000['Number'] + count_2000['Global_Sales']
print(count_2000.sort_values(by='Sum', ascending=False)[['Number','Global_Sales']].plot(kind='bar', stacked=True, title='Популярность жанров до 2000 года', color=colors))

count_2020 = table[table['Year'] > 2000].groupby('Genre').agg({'Name' : 'count', 'Global_Sales' : 'sum'})
count_2020.columns = ['Number', 'Global_Sales']
count_2020['Sum'] = count_2020['Number'] + count_2020['Global_Sales']
print(count_2020.sort_values(by='Sum', ascending=False)[['Number','Global_Sales']].plot(kind='bar', stacked=True, title='Популярность жанров после 2000 года', color=colors))
```

```python
# 2
count_by_year = table.groupby('Year').count()['Name']
print(count_by_year.plot(color=colors[1]))
```

```python
# 3
popular_publishers = table.groupby('Publisher').agg({'Name' : 'count'}).sort_values(by='Name', ascending=False).reset_index()['Publisher'][:3].to_list()
for i in range(3):
    print(table[(table['Publisher'] == popular_publishers[i])].groupby('Platform').agg({'Name' : 'count'}).plot(kind='bar', stacked=True, color=colors))
```

```python
# 4 1980-2000
print(table[(table['Year'] >= 1980) & (table['Year'] <= 2000)].
      agg({'Other_Sales' : 'sum', 'NA_Sales' : 'sum', 'EU_Sales' : 'sum', 'JP_Sales' : 'sum'}).
           plot(kind='pie', autopct='%1.0f%%', title='Объем продаж за 1980 - 2000 года', colors=colors))
```

```python
# 4 2001-2020
print(table[(table['Year'] >= 2001) & (table['Year'] <= 2020)].
      agg({'Other_Sales' : 'sum', 'NA_Sales' : 'sum', 'EU_Sales' : 'sum', 'JP_Sales' : 'sum'}).
           plot(kind='pie', autopct='%1.0f%%', title='Объем продаж за 2001 - 2020 года', colors=colors))
```
